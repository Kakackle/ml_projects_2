#### Function to display a random image from the dataset folder

```python
# View an image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random

def view_random_image(target_dir, target_class):
  # Setup target directory (we'll view images from here)
  target_folder = target_dir+target_class

  # Get a random image path
  random_image = random.sample(os.listdir(target_folder), 1)

  # Read in the image and plot it using matplotlib
  img = mpimg.imread(target_folder + "/" + random_image[0])
  plt.imshow(img)
  plt.title(target_class)
  plt.axis("off");

  print(f"Image shape: {img.shape}") # show the shape of the image

  return img
```

```python
# View a random image from the training dataset
img = view_random_image(target_dir="pizza_steak/train/",
                        target_class="pizza")
```

#### Data preprocessing
- scaling to [0,1]
- loading into batches
- reducing image size to (224,224)

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Set the seed
tf.random.set_seed(42)

# Preprocess data (get all of the pixel values between 1 and 0, also called scaling/normalization)
train_datagen = ImageDataGenerator(rescale=1./255)
valid_datagen = ImageDataGenerator(rescale=1./255)

# Setup the train and test directories
train_dir = "pizza_steak/train/"
test_dir = "pizza_steak/test/"

# Import data from directories and turn it into batches
train_data = train_datagen.flow_from_directory(train_dir,
                                               batch_size=32, # number of images to process at a time 
                                               target_size=(224, 224), # convert all images to be 224 x 224
                                               class_mode="binary", # type of problem we're working on
                                               seed=42)

valid_data = valid_datagen.flow_from_directory(test_dir,
                                               batch_size=32,
                                               target_size=(224, 224),
                                               class_mode="binary",
                                               seed=42)
```

#### First model

```python
model_1 = tf.keras.models.Sequential([
  tf.keras.layers.Conv2D(filters=10, 
                         kernel_size=3, # can also be (3, 3)
                         activation="relu", 
                         input_shape=(224, 224, 3)), # first layer specifies input shape (height, width, colour channels)
  tf.keras.layers.Conv2D(10, 3, activation="relu"),
  tf.keras.layers.MaxPool2D(pool_size=2, # pool_size can also be (2, 2)
                            padding="valid"), # padding can also be 'same'
  tf.keras.layers.Conv2D(10, 3, activation="relu"),
  tf.keras.layers.Conv2D(10, 3, activation="relu"), # activation='relu' == tf.keras.layers.Activations(tf.nn.relu)
  tf.keras.layers.MaxPool2D(2),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(1, activation="sigmoid") # binary activation output
])
```

```python
model_1.compile(loss='binary_crossentropy',
                optimizer=tf.keras.optimizers.Adam(),
                metrics=['accuracy'])
```

```python
model_1.fit(train_data, epochs=5,
            steps_per_epoch=len(train_data),
           validation_data=valid_data,
           validation_steps = len(valid_data))
```

#### Try a non-convolutional model

```python
model_2 = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(224,224,3)),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
model_2.compile(
    loss=tf.keras.losses.binary_crossentropy,
    optimizer=tf.keras.optimizers.Adam(),
    metrics=['accuracy',]
)
```

```python
history_2 = model_2.fit(train_data,
                        epochs=5,
                        steps_per_epoch=len(train_data),
                        validation_data = valid_data,
                        validation_steps = len(valid_data)
                       )
```

```python
history_2_df = pd.DataFrame(history_2.history)
```

```python
plt.figure(figsize=(10,7))
history_2_df.loc[:, ['loss', 'accuracy']].plot()
plt.figure(figsize=(10,7))
history_2_df.loc[:, ['val_loss', 'val_accuracy']].plot()
```

```python
def display_10_folder_images(folder_path, labels):
    """
    folder_path -  eg. './pizza_steak/train/'
    labels  - eg. ['pizza', 'steak']
    """
    plt.figure(figsize=(20,10))
    random_labels = random.sample(labels, counts=[5,5], k=10)
    for label_i, label in enumerate(random_labels):
        label_images = os.listdir(os.path.join(folder_path, label))
        image = random.sample(label_images, 1)[0]
        plt.subplot(2,5,label_i+1)
        img = mpimg.imread(os.path.join(folder_path, image))
        plt.imshow(img)
        plt.title(label)
        plt.axis("off");
```

```python
display_10_folder_images('./pizza_steak/train', ['pizza', 'steak'])
```
